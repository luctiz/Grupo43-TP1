```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv', dtype = {'piscina':'int32', 'gimnasio':'int32', 'usosmultiples':'int32', 'escuelascercanas':'int32',
                                         'centroscomercialescercanos':'int32', 'precio':'int32', 'id':'int32'},parse_dates = ['fecha'])
```

```python
df = df.replace(np.nan, 0)
```

```python
df = df.astype({'banos':'int32', 'antiguedad':'int32', 'garages':'int32', 'habitaciones':'int32', 'metroscubiertos':'int32', 'metrostotales':'int32',
                    'idzona':'int32'})
```

```python
categorias = df['tipodepropiedad']
categorias = categorias.astype('category').dtypes
df = df.astype({'tipodepropiedad':'category'})
```

Analisis de precios vs habitaciones, garages, ba√±os, metros cubiertos, metros totales.
Empiezo a analizar como cambia el precio segun la cantidad de habitaciones

```python
g = df[df['habitaciones']>0].groupby('habitaciones').mean().plot(kind='bar', y='precio', figsize=(12,7), fontsize=12, legend=False)
g.set_title("Precio promedio de propiedades seg√∫n cantidad de habitaciones", fontsize=22)
g.set_xlabel("Habitaciones", fontsize=15)
g.set_ylabel("Precio", fontsize=15)
plt.show()
```

```python
c = df[(df['habitaciones']>0) & (df['tipodepropiedad'] == "Casa")].groupby('habitaciones').mean()
a = df[(df['habitaciones']>0) & (df['tipodepropiedad'] == "Apartamento")].groupby('habitaciones').mean()
cc = df[(df['habitaciones']>0) & (df['tipodepropiedad'] == "Casa en condominio")].groupby('habitaciones').mean()
c['Casa']=c['precio']
c['Apartamento']=a['precio']
c['Casa en condominio']=cc['precio']
g = c.plot(kind='bar', y=['Casa', 'Apartamento', 'Casa en condominio'], figsize=(12,7), fontsize=12)
g.legend(fontsize=15)
g.set_title("Precio promedio de ciertas propiedades seg√∫n la cantidad de habitaciones", fontsize=22)
g.set_xlabel("Habitaciones", fontsize=15)
g.set_ylabel("Precio", fontsize=15)
plt.yticks([x for x in range(1000000,10000000,1000000)])
plt.show()
```
